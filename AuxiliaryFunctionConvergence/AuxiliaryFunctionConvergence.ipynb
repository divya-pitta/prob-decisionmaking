```python
import math
   
def f(x):
    return math.log(cosh(x))

def fprime(x):
    return (math.sinh(x)/math.cosh(x))

def fdprime(x):
    return (1/(cosh(x)*cosh(x)))

def qx_xn(x, xn):
    return (f(xn) + (x-xn)*fprime(xn) + 0.5*(x-xn)*(x-xn))

def qprime(x, xn):
    return (fprime(xn) + x - xn)
```

```python
def updateNewtonsMethod(x0):
    plotX = []
    plotY = []
    plotY.append(x0)
    plotX.append(0)
    x = x0
    count = 1
    while(True):
        plotX.append(count)
        xold = x
        x = (x - (math.sinh(x)*math.cosh(x)))
        plotY.append(x)
        if(abs(x-xold) <= 0.000001):
            break
    plt.plot(plotX, plotY)
    plt.show()
```

```python
updateNewtonsMethod(1)
```

```python
def g(x):
    res = 0
    for i in range(1, 11):
        res = res + math.log(math.cosh(x+(2/math.sqrt(i))))
    return res/10

def gprime(x):
    res = 0
    for i in range(1, 11):
        res = res + (math.sinh(x+(2/math.sqrt(i)))/math.cosh(x+(2/math.sqrt(i))))
    return res/10
```
